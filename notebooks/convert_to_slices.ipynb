```python
import os

os.mkdir("train_slice_ims")
os.mkdir("train_slice_gts")

os.mkdir("valid_slice_ims")
os.mkdir("valid_slice_gts")
```

```python
import numpy as np
import glob

# Get the paths
train_paths = np.array([os.path.basename(i).split('.')[0] for i in glob.glob(f'train/*.im')])
val_paths = np.array([os.path.basename(i).split('.')[0] for i in glob.glob(f'valid/*.im')])
```

```python
import h5py
import sys
import matplotlib.pyplot as plt
sys.path.append('../src')
from utils import crop_im, clip_and_norm
```

```python
for i in range(len(train_paths)):

    # get full paths and read in whole image
    im_path = os.path.join("train/", train_paths[i] + '.im')
    seg_path = os.path.join("train/", train_paths[i] + '.seg')
    with h5py.File(im_path,'r') as hf:
        image = np.array(hf['data'])
    with h5py.File(seg_path,'r') as hf:
        mask = np.array(hf['data'])

    #medial meniscus
    med_mask = mask[...,-1]

    #lateral
    lat_mask = mask[...,-2]

    #both together
    minisc_mask = np.add(med_mask,lat_mask)
    mask = np.clip(minisc_mask, 0, 1)

    # crop image/mask
    image = crop_im(image)
    mask = crop_im(mask)

    # normalise image
    image = clip_and_norm(image, 0.005)

    # get slice and save in folder
    for slice_num in range(160):
        im_slice = image[...,slice_num]
        mask_slice = mask[...,slice_num]

        np.save(f"train_slice_ims/{train_paths[i]}_slice_{slice_num}", im_slice)
        np.save(f"train_slice_gts/{train_paths[i]}_slice_{slice_num}", mask_slice)

    print(f"{i} images split")

```
